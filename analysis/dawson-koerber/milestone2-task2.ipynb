## Task 2. Method Chaining and writing Python programs

### Step 1: Build and test your method chain(s)

```python
import pandas as pd
import numpy as np

data = pd.read_csv("../../data/raw/insurance.csv") #Load dataframe

df1 = (
    data
    .drop('children' , 1)  #Drops the children
    .drop('region' , 1)    #Drops the region
    [data['bmi'] <= 45]    #Removes people with BMIs greater than 45
)

df2 = (
    df1
    .reset_index()          #Resets the index
    .drop('index' , 1)      #Removes the od index
)
```

### Step 2: Wrap your method chain(s) in a function

```python
def wranglemania2K21 (datapath):
    
    data = pd.read_csv(datapath) #Load dataframe

    df1 = (
        data
        .drop('children' , 1)  #Drops the children
        .drop('region' , 1)    #Drops the region
        [data['bmi'] <= 45]    #Removes people with BMIs greater than 45
    )

    df2 = (
        df1
        .reset_index()          #Resets the index
        .drop('index' , 1)      #Removes the od index
    )
    
    return df2                  #Returns the processed dataframe
```

```python
wranglemania2K21("../../data/raw/insurance.csv")
```
